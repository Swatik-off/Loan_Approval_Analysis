Does loan approval depend on applicant income?

Plot: Boxplot or Violin Plot (Income vs Loan_Status)

Purpose: Compare income levels of approved vs. rejected applications.

```python
df.head(5)
```

```python
df['loan_status'].unique()
```

```python
df1 = df[(df['loan_status'] =='Approved') | (df['loan_status'] =='Declined') ]
```

```python
 df1['loan_status'].map({'Declined':0, 'Approved': 1})
```

```python
df1['monthly_income'].describe()
```

we have to find if approval depends on persons income or not

```python
plt.boxplot([df1[df1['loan_status'] =='Approved']['monthly_income'], df1[df1['loan_status'] =='Declined']['monthly_income']], labels= ['Approved', 'Not Approved'])
plt.title("Applicant income vs Loan Approval")
plt.xlabel("Loan status")
plt.ylabel("Applicant Income")
plt.tight_layout()
plt.grid(axis = 'y', linestyle ='--', alpha = 0.5)
plt.show()
```

The boxplot shows that applicant income alone does not significantly distinguish between those whose loans were approved and those whose loans were not. Both categories exhibit similar medians and distributions, with wide income variability and many high-income outliers in both groups. This implies that loan approval decisions are likely influenced by additional factors beyond just the applicant's income — such as credit history, co-applicant income, loan amount requested, or past defaults.
